```python
from sklearn.feature_extraction.text import CountVectorizer

document = [
    'i love to eat pizza',
    'pizza is a junk food',
    'pizza is the best',
    'pasta i love to eat'
]

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(document)
```
