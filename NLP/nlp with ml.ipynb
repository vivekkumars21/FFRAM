```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.txt',sep= ";",header = None, names= ['text','emotion'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df['emotion'] = le.fit_transform(df['emotion'])
df
```

```python
df['text'] = df['text'].apply(lambda x : x.lower())
```

```python
import string

def remove_punch(txt):
    return txt.translate(str.maketrans(' ',' ',string.punctuation))
```

```python
df['text'] = df['text'].apply(remove_punch)
```

```python
def remove_digit(txt):
    new = ""
    for i in txt:
        if not i.isdigit():
            new = new + i
    return new

df['text'] = df['text'].apply(remove_digit)
```

```python
df
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
def remove(txt):
    words = word_tokenize(txt)
    return words

df['text'] = df['text'].apply(remove)
```

```python
df.loc[0]['text']
```

```python
stop_words = set(stopwords.words('english'))

def remove_word(txt):
    cleaned = []
    for i in txt:
        if not i in stop_words:
            cleaned.append(i)
    return cleaned

df['text'] = df['text'].apply(remove_word)
df['text'] = df['text'].apply(lambda x : " ".join(x))
```

```python
df.loc[1]['text']
```

```python
from sklearn.feature_extraction.text import TfidfTransformer,CountVectorizer

bow_vecterization = CountVectorizer()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['text'], df['emotion'], test_size=0.2, random_state=42)
```

```python
X_train_bow = bow_vecterization.fit_transform(X_train)
X_test_bow = bow_vecterization.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

model_nb = MultinomialNB()
```

```python
model_nb.fit(X_train_bow,y_train)
```

```python
y_pred = model_nb.predict(X_test_bow)
```

```python
accuracy_score(y_test,y_pred)
```
