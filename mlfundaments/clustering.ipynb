```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
X , y_true = make_blobs(n_samples= 500, centers= 3, random_state= 42,cluster_std=0.8)
```

```python
df = pd.DataFrame(X,columns=['feature1','feature2'])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
inertia = []
K_range = range(1,11)
```

```python
for k in K_range:
    kmean = KMeans(n_clusters= k,random_state=42)
    kmean.fit(X_scaled)
    inertia.append(kmean.inertia_)
```

```python
plt.plot(K_range,inertia,marker = 'o')
```

```python
kmean_final = KMeans(n_clusters= 3,random_state=42)
```

```python
cluster_label = kmean_final.fit_predict(X_scaled)
```

```python
cluster_label
```

```python
sns.scatterplot(x = df['feature1'],y = df['feature2'], hue = cluster_label)
```

```python
X, y_true = make_moons(n_samples=500,noise= 0.05,random_state=42)
```

```python
from sklearn.cluster import KMeans,DBSCAN
```

```python
df  = pd.DataFrame(X, columns=['feature1','feature2'])
```

```python
X_scaled_k = scaler.fit_transform(df)
```

```python
kmean = KMeans(n_clusters=2,random_state=42)
kmean_label = kmean.fit_predict(X_scaled_k)
```

```python
df['cluster'] = kmean_label
```

```python
sns.scatterplot(x = df['feature1'],y = df['feature2'], hue= df['cluster'])
```

```python
dbscan = DBSCAN(eps=0.3,min_samples=5)
db_label = dbscan.fit_predict(X_scaled_k)
```

```python
df['db_lab'] = db_label
```

```python
sns.scatterplot(x = df['feature1'], y = df['feature2'], hue=df['db_lab'])
```
