```python
class Student:
    def __init__(self,name,branch):
        self.name = name
        self.branch = branch

    def intro(self):
        print(f"Your name is {self.name} and your branch is {self.branch}")


s1 = Student("VIVEK", "IT")
s1.intro()
```
