# Data Visualization using Matplotlib, Seabon, Ploty


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x**2
print(x,y)
```

```python
plt.plot(x,y)
plt.title("Square Of Values")
plt.xlabel("Values")
plt.ylabel("Squares")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x**2,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(y**2,x)
```

```python
fig = plt.figure(figsize=(16,9),dpi=100)
axis1 = fig.add_axes([0.1,0.1,.8,.8])

axis1 = axis1.plot(x,y)
plt.title("hello")
plt.xlabel("Name")
axis2 = fig.add_axes([.3,.3,.5,.5])
axis2.plot(x,y)
axis2 = plt.xlabel("Velle")
```
