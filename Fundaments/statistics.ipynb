```python
import numpy as np
import seaborn as sns
```

```python
arr = np.array([13,343,6654,865,224,453])
print(np.mean(arr))
```

```python
x = np.mean(arr)
l = []
for i in arr:
    sigma = (i - x)**2
    l.append(sigma)
arr2 = np.array(l)
print(arr2)
total = np.sum(arr2)
print(total)

var = total/np.size(arr)
print(var)
std = np.sqrt(var)
print(std)
```
