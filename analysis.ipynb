```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random




arr1 = np.array([2,3,4,5,6,7,8])
arr2 = np.array([7,5,3,99,12,45,76])

c = np.array([
   random.randint(1,25) * random.choice(arr1) + random.choice(arr2)
    for i in range(5)
])

print(c)



plt.hist(c)
```
