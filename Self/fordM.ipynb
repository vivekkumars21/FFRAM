```python
X = data.drop(columns= ['price'], axis= 1)
y = data['price']
```

```python
X
```

```python
data.columns
```

```python
X_one_encoded = pd.get_dummies(X, columns=['model', 'transmission','fuelType'])
X_one_encoded = X_one_encoded.astype(int)
X_one_encoded
```

```python
from sklearn.preprocessing import LabelEncoder

encode = LabelEncoder()
columns = [ 'model', 'transmission','fuelType']

Xlabel = X.copy()

for i in columns:
    Xlabel[i] = encode.fit_transform(Xlabel[i])

Xlabel
```

```python
from sklearn.preprocessing import StandardScaler
cols = ['year','mileage','tax','engineSize','mpg']
scaler = StandardScaler()
X_one_encoded[cols] = scaler.fit_transform(X_one_encoded[cols])
X_one_encoded
```

```python
Xlabel[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax',
       'mpg', 'engineSize']] = scaler.fit_transform(Xlabel[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax',
       'mpg', 'engineSize']] )
Xlabel
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_one_encoded, y, train_size= 0.2, random_state= 42)
model = LinearRegression()

model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
r2 = r2_score(y_test,y_pred)
r2
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xlabel, y, train_size= 0.2, random_state= 42)
model2 = LinearRegression()

model2.fit(X_train,y_train)
```

```python
y_pred = model2.predict(X_test)
y_pred
```

```python
r2 = r2_score(y_test,y_pred)
r2
```
